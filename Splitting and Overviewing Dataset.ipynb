```python
import os


TRAIN_PATH = os.path.join("data", "train")
TEST_PATH = os.path.join("data", "test")
VAL_PATH = os.path.join("data", "val")

if not os.path.exists(TRAIN_PATH):
    os.mkdir(TRAIN_PATH)
    
if not os.path.exists(TEST_PATH):
    os.mkdir(TEST_PATH)

if not os.path.exists(VAL_PATH):
    os.mkdir(VAL_PATH)
```

```python
os.mkdir(os.path.join(TRAIN_PATH, "images"))
os.mkdir(os.path.join(TRAIN_PATH, "labels"))
os.mkdir(os.path.join(TEST_PATH, "images"))
os.mkdir(os.path.join(TEST_PATH, "labels"))
os.mkdir(os.path.join(VAL_PATH, "images"))
os.mkdir(os.path.join(VAL_PATH, "labels"))
```

```python
TRAIN_RATIO = len(os.listdir("data/images/"))*.7
TEST_RATIO = len(os.listdir("data/images/"))*.15
VAL_RATIO = len(os.listdir("data/images/"))*.15
```

```python
MAIN_FOLDER_IMAGES = os.path.join("data", "images")
MAIN_FOLDER_LABELS = os.path.join("data", "labels")

keys = [image.split(".")[0] for image in os.listdir(MAIN_FOLDER_IMAGES)]
```

```python
print(keys)
```

```python
dataset = []

for key in keys:
    image_path = os.path.join(MAIN_FOLDER_IMAGES, f"{key}.jpg")
    label_path = os.path.join(MAIN_FOLDER_LABELS, f"{key}.json")
    dataset.append([image_path, label_path])
```

```python
import random
import math
```

```python
random.shuffle(dataset)

TRAIN_DATA = dataset[:math.ceil(TRAIN_RATIO)]
TEST_DATA = dataset[math.ceil(TRAIN_RATIO):math.ceil(TRAIN_RATIO)+math.floor(TEST_RATIO)]
VAL_DATA = dataset[math.ceil(TRAIN_RATIO)+math.floor(TEST_RATIO):]
```

```python
import shutil

for td in TRAIN_DATA:
    img_td = td[0]
    lab_td = td[1]
    shutil.move(img_td, os.path.join("data", "train", "images"))
    shutil.move(lab_td, os.path.join("data", "train", "labels"))    
```

```python
import shutil

for td in TEST_DATA:
    img_td = td[0]
    lab_td = td[1]
    shutil.move(img_td, os.path.join("data", "test", "images"))
    shutil.move(lab_td, os.path.join("data", "test", "labels"))    
```

```python
import shutil

for td in VAL_DATA:
    img_td = td[0]
    lab_td = td[1]
    shutil.move(img_td, os.path.join("data", "val", "images"))
    shutil.move(lab_td, os.path.join("data", "val", "labels"))    
```

```python
import tensorflow as tf
import json
import numpy as np
from matplotlib import pyplot as plt
```

```python
images = tf.data.Dataset.list_files('data/images/*.jpg')
```

```python
images.as_numpy_iterator().next()
```

```python
def load_image(x): 
    byte_img = tf.io.read_file(x)
    img = tf.io.decode_jpeg(byte_img)
    return img
```

```python
images = images.map(load_image)
```

```python
images.as_numpy_iterator().next()
```

```python
type(images)
```

```python
image_generator = images.batch(4).as_numpy_iterator()
```

```python
plot_images = image_generator.next()
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(20,20))
for idx, image in enumerate(plot_images):
    ax[idx].imshow(image) 
plt.show()
```
